```python
filelist=['object_p_2006.csv','object_p_2007.csv', 'object_p_2008.csv','object_p_2009.csv','object_p_2010.csv','object_p_2011.csv','object_p_2012.csv','object_p_2013.csv','object_p_2014.csv','object_p_2015.csv',]
print(filelist[0])
#創一0矩陣
all_cloud=np.zeros((1,19))
#讀入檔案數量
file_num=1
#將所有資料讀入組成一完整矩陣
for i in range(0,file_num):
    data = pd.read_csv(filelist[i])
    sizes=len(data)
    print(sizes)
    cloud=np.empty((sizes,19))
    cloud[:,:]=data.iloc[0:sizes,:]
    all_cloud=np.vstack((all_cloud,cloud))
    print(all_cloud.shape)
```
